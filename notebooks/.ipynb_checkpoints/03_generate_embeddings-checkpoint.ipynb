```python
import pandas as pd

df = pd.read_csv("../data/processed/protein_sequences.csv")
df.head()
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("facebook/esm2_t6_8M_UR50D")
model = AutoModel.from_pretrained("facebook/esm2_t6_8M_UR50D")

```

```python
import numpy as np

embeddings = []

for idx, row in df.iterrows():
    seq = row["sequence"]

    inputs = tokenizer(seq, return_tensors="pt", truncation=True)

    with torch.no_grad():
        outputs = model(**inputs)

    # Mean pooling across amino acids
    emb = outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

    embeddings.append(emb)
```

```python
df["embedding_vector"] = embeddings
```

```python
df["embedding_vector"] = df["embedding_vector"].apply(lambda x: ",".join(map(str, x)))

df.to_csv("../data/processed/protein_embeddings.csv", index=False)
```
