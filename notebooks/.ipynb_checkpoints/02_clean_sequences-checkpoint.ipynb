```python
from Bio import SeqIO
import pandas as pd

records = list(SeqIO.parse("../data/raw/spike_SARS_CoV_2_spike_AND_reviewed_2026_03_27.fasta", "fasta"))
```

```python
data = []

for record in records:
    protein_id = record.id
    sequence = str(record.seq)
    length = len(sequence)

    # Try to extract protein name from description
    desc = record.description
    if "Spike" in desc or "spike" in desc:
        protein_name = "Spike glycoprotein"
    else:
        protein_name = desc

    data.append({
        "protein_id": protein_id,
        "sequence": sequence,
        "protein_name": protein_name,
        "virus": "SARS-CoV-2",
        "length": length
    })

df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv("../data/processed/protein_sequences.csv", index=False)
```
